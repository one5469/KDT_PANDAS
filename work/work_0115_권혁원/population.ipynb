```python
# 모듈 로드
import pandas as pd
import numpy as np
```

```python
# 파일 경로
filename = './시도별 전출입 인구수.xlsx'
```

```python
# 파일 불러오기
# 역시 뭔가 이상하다
excel = pd.read_excel(filename)
excel
```

```python
excel.info()
```

```python
# 아무래도 저 NaN부터 어떻게든 해야할듯
# 시리즈를 떼어보자
FromSR = excel['전출지별']
FromSR
```

```python
# nan 데이터 채우기
# 이건 아닌거 같은데
area = ''
areas = []
for v in FromSR.values:
    if str(v) == 'nan':
       areas.append(area)
    else:
        area = v
        areas.append(area)
excel['전출지별'] = areas
```

```python
excel
```

### 서울로 가장 많이 이주한 지역은?

```python
# 인덱스를 전입지로 변경하고 가져온다
toSeoul = excel.set_index('전입지별').loc['서울특별시'].set_index('전출지별')
toJeju = excel.set_index('전입지별').loc['제주특별자치도'].set_index('전출지별')
```

```python
toSeoul
```

```python
toJeju
```

```python
# 뒤집어
toSeoul = toSeoul.T
toJeju = toJeju.T
```

```python
# 그놈의 NaN xx 진짜
# 0으로 다채워버리자
toSeoul.replace('-', 0, inplace=True)
toJeju.replace('-', 0, inplace=True)
```

```python
# 총량을 구해보자
totalToSeoul = toSeoul.iloc[1:].astype('int64').sum()
totalToJeju = toJeju.iloc[1:].astype('int64').sum()
```

```python
# 서울로 가장 많이 이주한 지역은 경기도
totalToSeoul.sort_values(ascending=False)
```

```python
# 제주로 가장 많이 이주한 지역은 서울
totalToJeju.sort_values(ascending=False)
```
