```python
# 모듈 불러오기
import pandas as pd
import numpy as np
```

```python
# 파일 경로
filename = './남북한발전전력량.xlsx'
```

```python
# 파일 불러오기
# 이럴수가 그냥 받았더니 자료형이 개판이야
excel = pd.read_excel(filename)
excel
```

```python
# 그렇다면 보기좋게 이쁘게 다듬자
# 먼저 남북한을 나눈다
excel.set_index('전력량 (억㎾h)', inplace=True)
southDF = excel.iloc[0:5].set_index('발전 전력별')
southDF
```

```python
# 이제 좀 보기 좋네
# 얘넨 근데 핵으로 무기만 만드나?
northDF = excel.iloc[5:].set_index('발전 전력별')
northDF
```

```python
# 북한에 '신재생' 행을 추가
northDF.loc['신재생'] = 0
northDF
```

```python
# 총계를 추가해보자
# 먼저 행열을 전환하자
southDF = southDF.T
northDF = northDF.T
```

```python
southDF
```

```python
northDF
```

```python
# 계산전에 데이터타입을 바꾸자
# 빈 데이터는 0으로 간주
southDF['원자력'].replace('-', 0, inplace=True)
southDF['신재생'].replace('-', 0, inplace=True)
northDF['원자력'].replace('-', 0, inplace=True)
```

```python
for c in southDF.columns:
    southDF[c] = southDF[c].astype('int64')
for c in northDF.columns:
    northDF[c] = northDF[c].astype('int64')
```

```python
# sum 메소드를 이용해 합계 계산
southTotal = southDF.sum()
northTotal = northDF.sum()
print(southTotal, northTotal)
```

```python
# 총계를 추가해보자
southDF.loc['총계'] = southTotal.values
northDF.loc['총계'] = northTotal.values
```

```python
northDF
```

```python
# 다시 뒤집어보자
southDF = southDF.T
northDF = northDF.T
```
