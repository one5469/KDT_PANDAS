## [3장 코딩 챌린지]

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# 파일 경로
filename = '../../DATA/revolutionary_war.csv'
```

```python
# csv 파일 불러오기
rwDF = pd.read_csv(filename, parse_dates=['Start Date'])
rwDF
```

```python
# Start Date 컬럼으로 시리얼 생성
dateSR = rwDF['Start Date']
dateSR
```

```python
# 저장 형태 : Timestamp
dateSR[1]
```

```python
# NaN 데이터 삭제
dateSR.dropna(inplace=True)
dateSR.isna().sum()
```

```python
# 시리얼 재생성
dateSR = pd.Series([ d.strftime("%A") if d else None for d in dateSR ])
dateSR
```

```python
# 최종 출력은 요일을 인덱스 레이블로, 각 날짜(?)의 전투 횟수를 값으로 가지는 Series
# 이거도 번역이 이상한데
# 싸늘하다
dateSR.value_counts()
```

```python
# json 파일 불러오기
tvshDF = pd.read_json(filename2)
tvshDF
```

### [문제 1번]

```python
# json 정규화를 통해 DataFrame 생성
epiDF = pd.json_normalize(tvshDF['shows'], record_path='episodes', meta=['show', 'runtime', 'network'])
```

```python
epiDF
```

### [문제 2번]

```python
# 인덱스를 show로 설정
showDF = epiDF.set_index('show')
showDF
```

```python
# 인덱스 값을 이용해 DataFrame 분리
XFilesDF = showDF.loc['The X-Files']
LostDF = showDF.loc['Lost']
BuffyDF = showDF.loc['Buffy the Vampire Slayer']
```
