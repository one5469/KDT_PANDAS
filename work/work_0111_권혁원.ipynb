(1) 모듈 로딩

```python
import pandas as pd
```

(2) 데이터 준비

```python
dataFile = '../DATA/movies.csv'
```

(3) 데이터 저장

```python
movieDF = pd.read_csv(dataFile)
movieDF
```

(4) DataFrame의 속성들 출력

```python
att = movieDF.columns
print(list(att))
```

(5) DataFrame의 컬럼별로 따로 저장하기

```python
Rank = movieDF['Rank']
Title = movieDF['Title']
Studio = movieDF['Studio']
Gross = movieDF['Gross']
Year = movieDF['Year']

print(Rank, Title, Studio, Gross, Year, sep='\n\n')
```

(6) 컬럼별 데이터들의 속성 즉, index, values, shape, ndim, dtype 정보 출력

```python
cols = [Rank, Title, Studio, Gross, Year]

for c in cols:
    print('Column :', c.name)
    print(f'index : {c.index}\nvalues : {c.values}\nshape : {c.shape}\nndim : {c.ndim}\ndtype : {c.dtype}')
    print()
    print()
```

(7) 컬럼별 인덱스를 변경

```python
for c in cols:
    c.index = [ f'{c.name} {i}' for i in range(1, c.shape[0]+1) ]
    print(c)
    print()

```

(8) 컬럼별 데이터 즉 값 변경

```python
for c in cols:
    for i in range(c.shape[0]):
        origin = str(c[i])
        c[i] = origin + '.'
    print(c)
    print()
```

(9) DataFrame의 원소들 모두 출력 단, 2가지 방법 사용

```python
for row in movieDF.index:
    for col in movieDF.columns:
        print(movieDF.loc[row, col], end= ' // ')
```

```python
for row in range(movieDF.shape[0]):
    for col in range(movieDF.shape[1]):
        print(movieDF.iloc[row, col], end=' // ')
```
