```python
import pandas as pd
```

```python
filename = './titanic.csv'
titanic = pd.read_csv(filename)
titanic
```

```python
titanic.info()
```

```python
titanic.isna().sum()
```

```python
titanic.nunique()
```

```python
# 데이터 전처리
titanic.survived = titanic.survived.astype('category')
titanic.sex = titanic.sex.astype('category')
titanic.pclass = titanic.pclass.astype('category')
titanic['class'] = titanic['class'].astype('category')
titanic.embarked = titanic.embarked.astype('category')
```

```python
# embarked와 embark_town 데이터 확인
embark = titanic[['embarked', 'embark_town']]
embark
```

```python
for i in range(embark.shape[0]):
    if str(embark.iloc[i]['embarked']) != str(embark.iloc[i]['embark_town'])[0]:
        print(i)
```

```python
# 결과값 2개가 NaN임을 확인
embark.iloc[61], embark.iloc[829]
```

```python
# embark_town 컬럼 제거
titanic.drop(columns='embark_town', inplace=True)
titanic
```

```python
# 나이 결측치 제거
titanic.dropna(subset='age', inplace=True)
titanic
```

```python
# 전체 나이 통계값 분석
titanic['age'].describe(), titanic['age'].median()
```

```python
# 전체 나이 분포 분석
titanic['age'].unique()
```

```python
# 사망자 나이 분석
titanic[titanic['survived'] == 0]['age'].describe(), titanic[titanic['survived'] == 0]['age'].median()
```

```python
# 생존자 나이 분석
titanic[titanic['survived'] == 1]['age'].describe(), titanic[titanic['survived'] == 1]['age'].median()
```

```python
# 나이 범주화
titanic['agerange'] = [ (age//10)*10 for age in titanic.age ]
titanic['agerange'] = titanic['agerange'].astype('int32')
titanic
```

```python
# 나이별 생존자 비율 분석
titanic[titanic['survived'] == 0]['agerange'].value_counts().sort_index()
```

```python
titanic[titanic['survived'] == 1]['agerange'].value_counts().sort_index()
```

```python
# 남성 사망자 나이 분포
titanic[(titanic['survived'] == 0) & (titanic['sex'] == 'male')]['agerange'].value_counts().sort_index()
```

```python
# 여성 사망자 나이 분포
titanic[(titanic['survived'] == 0) & (titanic['sex'] == 'female')]['agerange'].value_counts().sort_index()
```

```python
# 남성 생존자 나이 분포
titanic[(titanic['survived'] == 1) & (titanic['sex'] == 'male')]['agerange'].value_counts().sort_index()
```

```python
# 여성 생존자 나이 분포
titanic[(titanic['survived'] == 1) & (titanic['sex'] == 'female')]['agerange'].value_counts().sort_index()
```
