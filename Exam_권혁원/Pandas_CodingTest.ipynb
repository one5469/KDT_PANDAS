[문1] 아래 조건을 만족하는 Series를 생성하세요.
1) 데이터 : 1~50 범위에서 7의 배수 정수
2) 인덱스 : 'a~z' 범위에서 데이터 갯수 만큼의 알파벳
3) 타  입 : 부호가 없는 1바이트 정수
4) 이  름 : sevens
5) 속  성 : index, values, dtype, ndim, shape, size

```python
values = [ i for i in range(7, 50, 7)]   # 1~50 범위의 7의 배수 정수
index = [ chr(i) for i in range(97, 97+len(values)) ]   # a~z범위 알파뱃 유니코드를 이용해 데이터 갯수만큼 리스트에 담음

sr1 = pd.Series(values, index=index, dtype='uint8', name='sevens')   # 시리즈 객체 생성
sr1
```

```python
sr1.info()       # 시리즈 객체 정보
```

```python
print(sr1.index, sr1.values, sr1.ndim, sr1.shape, sr1.size, sep='\n\n')      # 시리즈 객체 속성
```

[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
3) 결측치 처리 =>
   (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거
   (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
stock = pd.read_excel('./stock price.xlsx', index_col='id')
stock.isna().sum()
```

```python
# 결측치 제거 - 1개라도 NaN이면 다 제거
stock1 = stock.dropna()
```

```python
stock1.info()
```

```python
stock1.isna().sum()
```

```python
# 결측치 제거 - price가 NaN이면 제거
stock2 = stock.dropna(subset='price')
stock2.info()
```

```python
stock2.isna().sum()
```

```python
# 결측치(?) 제?거 - value가 10000이하면 제거
stock3 = stock[stock['value'] > 10000]
stock3.info()
```

```python
stock3[stock3['value'] <= 10000]
```

```python
# 결측치 대체 - 이전값
stock4 = stock.fillna(method='ffill')
stock4.isna().sum()
```

```python
stock4
```

```python
# 결측치 대체 - 이후값
stock5 = stock.fillna(method='bfill')
stock5.isna().sum()
```

```python
stock5
```

```python
# value가 1000이하 이거나 price가 90000이상인 데이터
stock6 = stock[(stock['value'] <= 1000) | (stock['price'] >= 90000)]
stock6.info()
```
