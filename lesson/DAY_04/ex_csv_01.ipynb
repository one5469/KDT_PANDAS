## [ csv 데이터 파일 다루기 ]

(1) 모듈 로딩

```python
import pandas as pd
```

(2) 데이터 준비

- 파일 이름 => 경로 + 파일명
- 경로 => 2가지 종류
    * 절대경로 : 드라이브부터 파일까지의 전체 경로 (ex:
    * 상대경로 : 현재 파일을 기준으로 경로 설정
         - ./  : 현재 위치 의미 (예: ./a.txt, a.txt => 현재 파일과 동일한 폴더/위치에 존재)
         - ../ : 한 단계 위로 의미 (예: ../Data/b.csv => 현재 파일이 존재하는 폴더 밖에 존재하는 파일)

```python
file = './test.csv'         # file = 'test.txt' 동일 의미
```

```python
file2 = '../DATA/movies.csv'
```

```python
file3 = './test2.csv'
```

(3) 데이터 저장 : CSV => DataFrame

```python
movieDF = pd.read_csv(file2)
```

(4) 데이터 확인

```python
# 전체적인 데이터 정보 확인 => info() 메소드
movieDF.info()
```

```python
# 실제 저장된 데이터 확인 => head(), tail()
print(movieDF.head(), movieDF.tail(), sep='\n\n')
```

(5) read_csv 파일 매개변수 살펴보기

- head 파라미터 : DataFrame의 컬럼명 설정하는 파라미터
    * 기본값 : csv 파일의 첫번째 라인의 데이터를 컬럼명으로 설정
    * None  : csv 파일의 첫번째 라인이 컬럼명 아님 의미 => 컬럼명 미지정
    * 행번호 : csv 파일의 행번호 라인이 컬럼명으로 설정, 그 이전 행번호 데이터는 사라짐!

```python
pd.read_csv(file2, header=None)
```

```python
pd.read_csv(file2, header=1)
```

```python
pd.read_csv(file2, header=5)
```

- index_col 파라미터 : 컬럼 중에서 행 인덱스로 설정, set_index()와 동일 역할

```python
pd.read_csv(file2, index_col='Rank')              # 단일 인덱스
```

```python
pd.read_csv(file2, index_col=['Rank', 'Title'])        # 멀티 인덱스
```

- sep / delimiter : csv 파일에서 데이터 구분자 설정
    * 기본값  : 쉼표(,)

```python
pd.read_csv(file, header=None, sep=';')
```

- skiprows : csv 파일에 읽어 들이지 않을 행 갯수를 지정하기 => 앞부분
- skipfooter : csv 파일에 읽어 들이지 않을 행 갯수를 지정하기 => 끝부분

```python
pd.read_csv(file)
```

```python
pd.read_csv(file3, sep=';', skiprows=3, header=None)
```

```python
pd.read_csv(file3, sep=';', skiprows=3, header=None, skipfooter=2, engine='python')
```

- usecols : 컬럼에서 특정 컬럼만 선택해서 로딩함
      * usecols = [0, 2, ...]
      * usecols = ['컬럼명1', '컬럼명2']

```python
pd.read_csv(file3, sep=';', skiprows=3, header=None, usecols=[0,2])
```

```python
pd.read_csv(file2, usecols=['Rank', 'Year'])
```

```python
pd.read_csv(file2, usecols=[0, 2])
```
