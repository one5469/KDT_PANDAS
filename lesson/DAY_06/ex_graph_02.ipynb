(2) 데이터 준비

### X, Y축 레이블 설정 및 눈금 설정
- x y축 데이터에 대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태 설정 => plt.xticks(), plt.yticks(), plt.tick_param()

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', [d+10000 for d in y], 'g^')
plt.title('[테스트]')

# x, y축 레이블
plt.xlabel("값", loc='right', fontsize='xx-large', color='red', mouseover=True)
plt.ylabel("결과", loc='top', fontsize='xx-large', color='green', rotation='horizontal')

# x, y축 눈금 설정
plt.xticks(ticks=[11, 12, 13, 14, 15], labels=['A', 'B', 'C', 'D', 'E'])
plt.yticks(color='blue')
```

### 범례 설정
- 여러개의 그래프를 표시하는 경우 그래프별 식별을 위해서 사용

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', [d+10000 for d in y], 'g^', label='결과')
plt.title('[테스트]')

# x, y축 레이블
plt.xlabel("값", loc='right', fontsize='xx-large', color='red', mouseover=True)
plt.ylabel("결과", loc='top', fontsize='xx-large', color='green', rotation='horizontal')

# x, y축 눈금 설정
plt.xticks(ticks=[11, 12, 13, 14, 15], labels=['A', 'B', 'C', 'D', 'E'])
plt.yticks(color='blue')

# 범례 설정
plt.legend(['red 결과', 'green 결과'], loc=0)
```

### 텍스트 추가
- 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text(시작x, 시작y, 출력텍스트)

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', [d+10000 for d in y], 'g^', label='결과')
plt.title('[테스트]')

# x, y축 레이블
plt.xlabel("값", loc='right', fontsize='xx-large', color='red', mouseover=True)
plt.ylabel("결과", loc='top', fontsize='xx-large', color='green', rotation='horizontal')

# x, y축 눈금 설정
plt.xticks(ticks=[11, 12, 13, 14, 15], labels=['A', 'B', 'C', 'D', 'E'])
plt.yticks(color='blue')

# 범례 설정
plt.legend(['red 결과', 'green 결과'], loc=0)

# 그래프 안에 텍스트 추가
plt.text(2, 4000, 'Happy New Year', backgroundcolor='yellow', color='red')

box={'facecolor':'.75', 'edgecolor':'k', 'boxstyle':'round', 'alpha':0.35}

plt.text(2, 6000, 'Happy New Year', bbox=box)
```

### 화살표 설정
- 그래프에 특정 지점에 화살표 표기
- text와 함께 종종 사용됨
- plt.annotatate()

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', [d+10000 for d in y], 'g^', label='결과')
plt.title('[테스트]')

# x, y축 레이블
plt.xlabel("값", loc='right', fontsize='xx-large', color='red', mouseover=True)
plt.ylabel("결과", loc='top', fontsize='xx-large', color='green', rotation='horizontal')

# x, y축 눈금 설정
plt.xticks(ticks=[11, 12, 13, 14, 15], labels=['A', 'B', 'C', 'D', 'E'])
plt.yticks(color='blue')

# 범례 설정
plt.legend(['red 결과', 'green 결과'], loc=0)

# 화살표 설정
plt.annotate('Hmm', xytext=(2, 3000), xy=(2.5, 400), arrowprops={'facecolor':'black', 'shrink':0.1})

# 격자 설정
plt.grid('both')
```
