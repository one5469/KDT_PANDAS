## [데이터 시각화 맛보기]

(1) 모듈 로딩

```python
import matplotlib.pyplot as plt
import pandas as pd
```

(2) 데이터 준비

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = range(100)
y = [ d**2 for d in x ]
```

(3) 그래프 그리기

- [기본] 라인/선 그래프
    * 시간의 변화를 표현해주는 그래프
    * 연속적/순서 있는 데이터를 시각화할 경우 주로 사용됨

```python
# plot(data) => y축 데이터로 사용
# plot(data1, data2) => x축에 data1, y축에 data2 데이터로 사용
# 반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환됨
ret = plt.plot(x, y)
print(type(ret), len(ret), type(ret[0]))

# 파이썬에서 표준출력 기능 내장 함수 => print() 동일 역할
plt.show()
```

```python
ret[0].get_xdata()
```

```python
x = range(1, 100, 10)
y = [ d**2 for d in x ]
```

- 라인/선 그래프에서 형식 즉 포맷지정 => str 타입의 포맷 설정
    * '색상형태선스타일' ==> 'ro--' : red circle dot-line
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b - blue, k = black
    * 대표적으로 많이 사용되는 형태 : o - 동그라미, ^ - 삼각형, < - 삼각형, > - 삼각형, v - 삼각형
    * 대표적으로 많이 사용되는 선스타일 : - 실선, -- 점선

```python
ret = plt.plot(x, y, 'ro')
```

```python
ret = plt.plot(x, y, 'ro--')
```

```python
ret = plt.plot(x, y, 'r^--')
```

```python
ret = plt.plot(x, y, 'r^--', x, [d+10000 for d in y], 'go--')
print(len(ret))
```
