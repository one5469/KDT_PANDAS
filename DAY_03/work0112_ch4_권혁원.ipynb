### 문제 1번

```python
# 날짜/시간...?
# 문제가 이상하다
import pandas as pd

filename = '../DATA/nfl.csv'
nfl = pd.read_csv(filename)
nfl
```

```python
# 책에 나온 풀이
nfl_ = pd.read_csv(filename, parse_dates = ['Birthday'])

nfl_
```

```python
# 매개변수를 어캐 다 암..?
# 그냥 바꾸기
def parser(birth):
    dates = birth.split('/')
    return dates[-1] + '-' + f'{dates[0]:0>2}' + '-' + f'{dates[1]:0>2}'

birth = nfl['Birthday']
nfl['Birthday'] = [ parser(birth) for birth in birth.values ]
print(nfl)
```

### 문제 3번

```python
# 책에 나온 방법
teamSr = nfl['Team']
print(teamSr.value_counts())
```

```python
# 완전탐색 풀이
# 메소드 모르면 뭘 할 수 있는데
teams_dict = dict(zip(teamSr.unique(), [ 0 for i in teamSr.unique()]))     # 팀 명 => key, 선수 명수 => value

for team in nfl.set_index('Team').index:     # 인덱스를 'Team'으로 바꾼 DataFrame에서 완전탐색
    teams_dict[team] += 1                    # 아니 이건 좀

print(teams_dict)
```
